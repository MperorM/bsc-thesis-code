```python
import json

with open("./models/classical_benchmarks.json", "r") as f:
    val_acc = json.loads(f.read())
```

```python
import matplotlib.pyplot as plt

plt.plot(val_acc["random_forest"], label="Random Forest")
plt.plot(val_acc["decision_tree"], label="Decision Tree")
plt.plot(val_acc["lstm_dropout"], label="lstm with dropout")
plt.plot(val_acc["lstm_no_dropout"], label="lstm no dropout")
plt.legend()
plt.ylabel("accuracy")
plt.gca().invert_xaxis()
plt.xlabel("timesteps until gameend")
```
